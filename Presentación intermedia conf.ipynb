## Pregunta de investigación #1: ¿La variable 'age' tiene algún tipo de distribución?

```python
plt.title('Histograma de Age', fontsize=20)
sns.distplot(df.age)
plt.show()
```

```python
age_hist=df['success'].plot(kind='hist')
plt.ylabel('Frequency',fontsize=17)
```

```python
sns.swarmplot(x='success',
              y='age',
              data=df)
plt.title('Success vs Age')
```

```python
sns.boxplot(x=df['success'],y=df['age'],whis=(0,100)).set_title('Box plot of success by age',fontsize=20)
```

```python
sns.lmplot(x='success',y='age',data=df, fit_reg= False)
plt.title('Scatterplot of Wingspan vs CWDistace', fontsize=20)
```

```python
sns.boxplot(x = df['age'],
            whis = (0,100)
           ).set_title('Box plot ')
```

```python
df.plot(kind='line')
plt.xlabel('id de las personas',fontsize=17)
plt.xticks(fontsize=17,rotation=45)
plt.yticks(fontsize=17, rotation=45)
plt.title('Line graph age and sucess',fontsize=20)
```
