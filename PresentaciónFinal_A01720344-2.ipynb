```python
df=pd.read_csv('classsification1.csv')
df.head()
```

## Descripción de variables
### La variable de 'Genero' es categórica o cualitativa, pues muestra si la persona es hombre o mujer . 
### Por el contrario, las variables de 'Edad','Interes'y'Exito' son numéricas o cuantitativas porque muestran datos representados por números. 

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('ggplot')
print(plt.style.available)
```

## Pregunta de Investigación 1: ¿Existe alguna correlación fuerte entre las varibles de interés y edad?

```python
sns.lmplot(x='Edad',y='Interes',data=df, fit_reg= False)
plt.title('Edad vs Interes', fontsize=20)
```

## Respuesta: 
#### La correlación entre ambas variables no es tan fuerte, pues los datos se encuentran dispersos entre ellos. Los datos tienen tendencia a ir en aumento, pero al llegar a la edad de 30 se empieza a disminuir tal interés. No se observan datos atípicos que puedan estar causando ruido en el gráfico. 

## Pregunta de Investigación 2: ¿Existe algun rángo de edad en específico en el que se logren más éxitos?

```python
sns.boxplot(x=df['Exito'],y=df['Edad'],whis=(0,100)).set_title('Box plot exito y edad',fontsize=20)
```

## Respuesta:
#### Al analizar los diagramas de caja y bigotes entre los éxitos, se observa como en la caja de éxitos (caja azul) la mediana se encuentra entre los 25 años. Al mimo tiempo, sus cuartiles tienen como límite superior 33 años e inferior 18 años aproximadamente. En conclusión, el rango de edades entre los 18 y 33 años es donde ambos géneros alcanzan el mayor número de éxitos. 

## Pregunta de Investigación 3: ¿Hay más hombres que mujeres interesdos en la clasificación del torneo de tennis?

```python
sns.swarmplot(x='Genero',
              y='Interes',
              data=df)
plt.title('Genero vs Interes')
```

## Respuesta:
#### No, los hombres tienen un índice de interes menor al de las mujeres. Al máximo que llega el interés de algun hombre es al 90%, mientras que un mayor número de mujeres si llega al 99% de interés. En ambos géneros, el interés se concentra mayormente en los rangos de 20-80%. 

## Pregunta de Investigación 4: ¿Se encuentran más mujeres adultas o hombres adultos dentro de la clasficación?

```python
sns.catplot(x='Edad',
            data=df,
            kind='count',
            hue='Genero'
           )
```

## Respuesta: 
#### Se encuentran más mujeres adultas a comparación de los hombres. Al analizar la gráfica, se muestra que casi todo el conteo de los hombres se concentra en las edades más jovenes. Por lo contrario, las edades de las mujeres estan más concentradas en las edades intermedias o más grandes. 

## Pregunta de Investigación 5: ¿Que género obtuvo más exito en los partios anteriores?

```python
fig=plt.figure()
ax1=fig.add_subplot(1,2,1) #un renglon, dos columnas, first position
sns.distplot(df.loc[df['Genero']=='F',
                        ['Exito']
                        ]
           )
ax2= fig.add_subplot(1,2,2) ##un renglon, dos columnas, second position
sns.distplot(df.loc[df['Genero']=='M',
                        ['Exito']
                       ]
           )
ax1.title.set_text('Genero F')
ax2.title.set_text('Genero M')
```

## Respuesta:
#### En la base de datos, se tiene a más mujeres registradas.Por esa razón, se tienen más datos sobre los éxitos de las mujeres. En este caso, se tiene un mayor número de victorias registradas de las mujeres en comparación las de los hombres. Los hombres tienen un promedio de 1.1 en victorias, y las mujeres 2.25. En conclusión, las mujeres han obtenido mayores éxitos.
