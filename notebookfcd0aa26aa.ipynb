```python
import pandas as pd
import numpy as np 
import seaborn as sns
```

```python
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
```

```python
df=pd.read_csv('../input/miccsv/mic.csv')
df.head()
```

```python
age_gender=pd.pivot_table(index=['Age','Sex'],
                         data=df,
                         aggfunc=len,
                         values='Availability')
age_gender=age_gender.rename(columns={'Availability':'Count'},)
age_gender=age_gender.reset_index()
age_gender
```

```python
fig1=sns.catplot(x='Age',kind='count',hue='Sex',data=df)
```

```python
fig2=px.sunburst(age_gender,path=['Age','Sex'],values='Count',)
fig2.show()
```

```python
fig2=px.sunburst(age_gender,path=['Age','Sex'],values='Count',
                 title='Age group por g√©nero de respondentes',
                 )
plt.savefig('/kaggle/working/age_group.png')
fig2.show()
```
