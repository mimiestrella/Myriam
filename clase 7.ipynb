```python
df= pd.read_excel('mic.xlsx',engine='openpyxl')
df.head()
```

```python
counts=df['Age'].value_counts()
type(count)
counts=counts.to_frame()
type(counts)
counts
counts['%']=(counts.iloc[:,:]/counts.loc[:,'Age'].sum(axis=0))*100
counts


#counts['%']
```

```python
counts=counts.reindex(['<=20','21 - 30','31 - 40','41 - 50'])
counts.reset_index(inplace=True)
counts
```

```python
counts=counts.rename({'index': 'Age_group'},axis=1)
```

```python
counts
```

```python
explode_list=[0.01,0.02,0.03,0.04] #ditancia para que se depare el pie
color_list=['gold','yellowgreen','coral','skyblue']
counts['Age'].plot(kind='pie',
                   figsize=(15,6),
                   startangle=90,
                   shadow=True,
                   labels= None,
                   autopct='%1.1f%%',
                   pctdistance=1.1,
                   explode=explode_list,
                   colors=color_list,
                  )
plt.legend(labels=counts.Age_group,
           bbox_to_anchor=(1.0,1.0)
          )
plt.title('Porcentaje de respondentes de diferentes grupos de edad',fontsize=20)
plt.show()
```

```python
x=list(counts['Age_group'])
y=list(counts['Age'])
plt.bar(x,y,color='pink')
plt.title('Numero de respondentes de diferentes grupos de edades',fontsize=20)
plt.show()
```

```python
age_genero_pivot=pd.pivot_table(index=['Age','Sex'],
                                aggfunc=len,
                                data=df, values='Availability')
age_genero_pivot
```

```python
print(age_genero_pivot.columns)
age_gender=age_genero_pivot.rename(columns={'Availability':'Count'}
                                  )
age_gender['%']=(age_gender.iloc[:,0]/age_gender.iloc[:,0].sum(axis=0))*100
age_gender
```
