Pregunta de investigación 1: Hay correlación entre wingspan con CWdistance?

```python
df=pd.read_csv('Cartwheeldata.csv')
df.head()
```

```python
import seaborn as sns
```

```python
sns.lmplot(x='Wingspan',y='CWDistance',data=df, fit_reg= False ,hue='Gender')
plt.title('Scatterplot of Wingspan vs CWDistace', fontsize=20)
```

```python
sns.swarmplot(x='Gender',
              y='CWDistance',
              data=df)
plt.title('Swarm plot of gender vs CWDistance')
```

```python
df.describe()
```

```python
sns.boxplot(data=df.loc[:,['Age','Height','Wingspan','CWDistance']])
plt.xticks(fontsize=17,rotation=45)
plt.yticks(fontsize=17)
plt.title('Boxplots', fontsize=17)
```

```python
sns.boxplot(data=df.loc[df['Gender']=='F',['Age','Height','Wingspan','CWDistance']])
plt.xticks(fontsize=17,rotation=45)
plt.yticks(fontsize=17)
plt.title('Boxplots', fontsize=17)
```

# Conclusion 
Al estudiar las variables no se ve ninguna correlación 

```python
fig=plt.figure()
ax1=fig.add_subplot(1,2,1) #un renglon, dos columnas, first position
sns.boxplot(data= df.loc[df['Gender']=='F',
                        ['Wingspan']
                        ]
           )
ax2= fig.add_subplot(1,2,2) ##un renglon, dos columnas, second position
sns.boxplot(data=df.loc[df['Gender']=='M',
                        ['Wingspan']
                       ]
           )
ax1.title.set_text('Genero F')
ax2.title.set_text('Genero M')
```
