```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("antropometria-dataset-2.csv")
```

```python
data_queretar = data[data['desc_ent']== "QUERETARO"]
```

```python
data_queretar.peso.mean()
```

```python
data_jal_sex1 = data[data['sexo']==1]
data_jal_sex1 = data_jal_sex1[data_jal_sex1['desc_ent']=="JALISCO"]
data_jal_sex1.peso.median()
```

```python
pd.pivot_table(data,index=["desc_ent","sexo"],values=["peso"], aggfunc=[np.mean]).head(60)
```
